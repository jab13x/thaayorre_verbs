## <font color='darkorange'>**Basics of Thaayorre verb grammar**</font>

Thaayorre <font color='darkturquoise'>**verb stems**</font> may be either simplex (consisting solely of a verb root) or complex. Complex verb stems may be formed either by compounding a (reduced) person part noun with a verb root, or by suffixing one or a more derivational suffixes to the verb root. **This grammar focuses only on simplex verb stems and the second type of complex (the one with derivational suffixes).**

Thaayorre verbs can be conjugated in nine different <font color='darkturquoise'>**tenses**</font>. For practical reasons, the reduplicated verb tense is not included in this grammar. Thus, **the eight verb tenses this grammar entails are:**

* **Nonpast:** `[NPST]` in the code.
* **Past perfective:** `[PPFV]` in the code.
* **Past imperfective:** `[PIMP]` in the code.
* **Imperative:** `[IMP]` in the code.
* **Subjunctive:** `[SUBJ]` in the code.
* **Counterfactual:** `[COUNTER]` in the code.
* **Purposive:** `[PURP]` in the code.
* **Imminence:** `[IMMIN]` in the code.

These tenses are created based on two main <font color='darkturquoise'>**types of conjugation**</font> (`[Conj1]` and `[Conj2]` in the code). Since one of the main aims of this grammar is to show the implementation of general alternation rules, **the conjugation of irregular verbs is not included**.

As mentioned, <font color='darkturquoise'>**simplex verb roots**</font> are conjugated using the tense suffixes (explained below). <font color='darkturquoise'>**Complex verb roots**</font> are created using derivational suffixes prior to conjugating them by attaching the tense suffixes.

These <font color='darkturquoise'>**derivational suffixes**</font> follow a specific order.

1. **Valency mark:** `[Val]` in the code.
2. **Reflexive or reciprocal mark:** `[Reflex]` or `[Recipr]` in the code.
3. **Motion mark:** `[Go]` or `[Run]`  in the code.

Other derivational suffixes can be found in Thaayorre verbs after the tense suffixes (e.g., first and second enclitic pronouns, and a range of pragmatic markers). These **post-tense suffixes are not included in this grammar**. It is worth noting that no examples have ever been recorded illustrating the combination of more than three derivational morphemes in Thaayorre verbs. This is why **this grammar focuses on the three main types** mentioned above. The only exception to this linear order is the combination of the valence increasing morpheme with the reflexive and reciprocal suffixes, as they may occur in either order. For simplicity, **the order shown above is the only one demonstrated in this grammar**.

In the case of **simplex verb roots**, the verbs randomly belong to one of the two conjugations. <font color='darkturquoise'>**The verbs included for `[Conj1]` in this grammar are:**</font>

* **kal** (carry)
* **katp** (hold)
* **mungk** (eat)
* **patp** (camp)
* **path** (bite)
* **pinirm** (dream)
* **reek** (give)
* **rint** (cook)
* **thunp** (throw)


<font color='darkturquoise'>**The verbs included for `[Conj2]` in this grammar are:**</font>

* **kerp** (finish)
* **kooc** (bark)
* **rirk** (rise, do)
* **therk** (return)
* **thowol** (play)
* **wene** (become, want)
* **wont** (fall)
* **yik** (say)
* **yoongke** (hang)

In spite of these original allocations, when a verb root becomes **complex** by the addition of suffixes, <font color='darkturquoise'>**the assignation of verb roots to conjugation classes is lexically-determined following these rules:**</font>

* **All verb stems derived by suffixation of the valence mark belong to Conjugation 1**, regardless of the conjugation of the root from which they are derived. The suffix used for valence is `-n`, but it can be `-ni` when followed by another suffix `-rr`. This results in the `['val_nirr']` alternation rule in the code.
* **All verb stems derived by suffixation of the reflexive and reciprocal marks belong to Conjugation 2**, regardless of the conjugation of the root from which they are derived. Based on the existing data, it does not seem to be possible to combine the reflexive and reciprocal morphemes in a single verb. The suffix used for reflexivity is `-e`, and the suffix used for reciprocity is `-rr`. When combined with valency, this type of suffixation overrides the former, so the verb goes to Conjugation 2.
* **Associated motion derivation has no effect on conjugation**, with the derived stem retaining the original conjugation of the simplex stem. This derivation can be used to mark that the event described by the verb root is accompanied by some kind of motion (typically a change in location). Thaayorre possesses two associated motion morphemes, both of which can express aspectual information as well as associated motion proper (i.e. that the event described by the verb root also involves some change of place). Nevertheless, the precise semantic categories they encode differ. All the linguistic and historical reasons behind these differences are out of the scope of this grammar. The suffix used for the `run` motion is `-ic`, but it can be `-ica` when followed by a consonant. This results in the `['run_ica']` alternation rule in the code. The suffix used for the `go` motion is `-nha`.

Finally, <font color='darkturquoise'>**tense suffixes**</font> are added either to the simplex verb roots or to the complex ones. <font color='darkturquoise'>**The suffixes for `[Conj1]` verbs are:**</font>

* **Nonpast:** `-r`
* **Past perfective:** `-V-rr` (see Note 1 below)
* **Past imperfective:** `-m` (see Note 2 below)
* **Imperative:** `-ø`
* **Subjunctive:** `-nh`
* **Counterfactual:** `-nhata`
* **Purposive:** `-nhatha`
* **Imminence:** `-rrth` (see Note 3 below)

<font color='darkturquoise'>**The suffixes for `[Conj2]` verbs are:**</font>

* **Nonpast:** `-ø`
* **Past perfective:** `-r`
* **Past imperfective:** `-m` (see Note 2 below)
* **Imperative:** `-ø`
* **Subjunctive:** `-nh`
* **Counterfactual:** `-nhata`
* **Purposive:** `-nhatha`
* **Imminence:** `-rrtha`

The combination of these conjugation suffixes with the derivational suffixes mentioned above means that, in some verb combinations, there could be three or even four consecutive `r`s. For instance, a verb showing reciprocity (`-rr`) in the past perfective of Conjugation 2 (`-r`) would have three final `r`. When this happens, only two `r`s (i.e., `-rr`) are kept. This results in the `['three_r']` and `['four_r']` alternation rules in the code.

*Note 1: The vowel used to create the past perfective in `[Conj1]` is completely random. For instance, the root `kal` adds an `i` to become `kalirr`, the root `patp` adds an `i` to become `patpirr`, the root `path` adds an `a` to become `patharr`, and the root `reek` adds an `a` to become `patharr`. This randomness results in the `['i_insertion']` and `['a_insertion']` alternation rules in the code.* <br>
*Note 2: The past imperfective can show an optional `nha` prior to the `m` suffix (i.e., `-nham`). Since it is optional, it is not included in this grammar.* <br>
*Note 3: A final `-a` here is optional, but not added in the rules to differentiate it from `[Conj2]` verbs, which always have it.*

## <font color='darkorange'>**Code**</font>

```python
from pyfoma import *
from pyfoma import Paradigm
from pyfoma import FST
```

```python
# Creating a right-linear grammar with input and output.
Grammar = {}
Grammar["S"] = [("", "Verb")]
Grammar["Verb"] = [("kal", "BaseC1"), ("katp", "BaseC1"), ("mungk", "BaseC1"), ("patp", "BaseC1"), ("path", "BaseC1"), \
                   ("pinirm", "BaseC1"), ("reek", "BaseC1"), ("rint", "BaseC1"), ("thunp", "BaseC1"), \
                   ("kerp", "BaseC2"), ("kooc", "BaseC2"), ("rirk", "BaseC2"), ("therk", "BaseC2"), ("thowol", "BaseC2"), \
                   ("wene", "BaseC2"), ("wont", "BaseC2"), ("yik", "BaseC2"), ("yoongke", "BaseC2")]

Grammar["BaseC1"] = [(("'[V]'", ""), "Conj1"), (("'[V]'", ""), "Deriv"), (("'[V]'", ""), "AssMotC1")]
Grammar["BaseC2"] = [(("'[V]'", ""), "Conj2"), (("'[V]'", ""), "Deriv"), (("'[V]'", ""), "AssMotC2")]

Grammar["Deriv"] = [(("'[Val]'", "+n"), "Conj1"), (("'[Val]''[Reflex]'", "+n+e"), "Conj2"), (("'[Val]''[Recipr]'", "+n+rr"), "Conj2"), \
                    (("'[Reflex]'", "+e"), "Conj2"), (("'[Recipr]'", "+rr"), "Conj2"), \
                    (("'[Val]'", "+n"), "AssMotC1"), (("'[Val]''[Reflex]'", "+n+e"), "AssMotC2"), (("'[Val]''[Recipr]'", "+n+rr"), "AssMotC2"), \
                    (("'[Reflex]'", "+e"), "AssMotC2"), (("'[Recipr]'", "+rr"), "AssMotC2")]

Grammar["AssMotC1"] = [(("'[Run]'", "+ic"), "Conj1"), (("'[Go]'", "+nha"), "Conj1")]
Grammar["AssMotC2"] = [(("'[Run]'", "+ic"), "Conj2"), (("'[Go]'", "+nha"), "Conj2")]

Grammar["Conj1"] = [(("'[Conj1]''[NPST]'", "+r"), "#"), \
                     (("'[Conj1]''[PPFV]'", "+rr"), "#"), \
                     (("'[Conj1]''[PIMP]'", "+m"), "#"), \
                     (("'[Conj1]''[IMP]'", ""), "#"), \
                     (("'[Conj1]''[SUBJ]'", "+nh"), "#"), \
                     (("'[Conj1]''[COUNTER]'", "+nhata"), "#"), \
                     (("'[Conj1]''[PURP]'", "+nhatha"), "#"), \
                     (("'[Conj1]''[IMMIN]'", "+rrth"), "#")]
Grammar["Conj2"] = [(("'[Conj2]''[NPST]'", ""), "#"), \
                     (("'[Conj2]''[PPFV]'", "+r"), "#"), \
                     (("'[Conj2]''[PIMP]'", "+m"), "#"), \
                     (("'[Conj2]''[IMP]'", ""), "#"), \
                     (("'[Conj2]''[SUBJ]'", "+nh"), "#"), \
                     (("'[Conj2]''[COUNTER]'", "+nhata"), "#"), \
                     (("'[Conj2]''[PURP]'", "+nhatha"), "#"), \
                     (("'[Conj2]''[IMMIN]'", "+rrtha"), "#")]
```

```python
# Creating and visualizing the lexicon based on the grammar.
fsts = {}
fsts['Lexicon'] = FST.rlg(Grammar, "S").epsilon_remove().determinize().minimize()
fsts['Lexicon'].view()
```

```python
# Viewing the lexicon.
print(Paradigm(fsts['Lexicon'], ".*"))
```

```python
# Defining consonants.
fsts['C'] = FST.re("[a-z]-[aeiou]")

# Alternation rules.
fsts['val_nirr'] = FST.re("$^rewrite(n:(ni) / \+ _ \+ (rr))")
fsts['run_ica'] = FST.re("$^rewrite((ic):(ica) / _ \+ ($C))", fsts)
fsts['i_insertion'] = FST.re("$^rewrite('':i / [lpmt] _ \+ (rr) #)")
fsts['a_insertion'] = FST.re("$^rewrite('':a / [hk] _ \+ (rr) #)")
fsts['four_r'] = FST.re("$^rewrite((rr):'' / _ \+ (rr))")
fsts['three_r'] = FST.re("$^rewrite((rr):'r' / _ \+ 'r')")
fsts['cleanup'] = FST.re("$^rewrite('+':'')")

# Composing and viewing the lexicon and the rules in a FST.
fsts['Thaayorre'] = FST.re("$Lexicon @ $val_nirr @ $run_ica @ $i_insertion @ $a_insertion @ $four_r @ $three_r @ $cleanup", fsts)
print(Paradigm(fsts['Thaayorre'], ".*"))
```
